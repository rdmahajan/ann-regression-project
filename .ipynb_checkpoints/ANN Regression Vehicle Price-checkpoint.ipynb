# Problem Statement
**Predict the resale price of vehicle based on it's characteristics.**

```python
# Install Dependencies
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as pp

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input, Dropout
from tensorflow.keras import regularizers
```

# Data understanding & Exploration

```python
# Load Dataset
df = pd.read_csv("ann_regression_vehicle_price.csv")
df.head()
```

```python
df.describe()
```

# Data Preprocessing

```python
# Split features & target
x = df.drop("resale_price", axis = 1)
y = df["resale_price"]
```

```python
# Train/Test Split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
# Features Scaling
ss = StandardScaler()
x_train = ss.fit_transform(x_train)
x_test = ss.transform(x_test)
```

# Modelling

```python
# Define NN

model = Sequential()

model.add(Input(shape = (x_train.shape[1],)))
model.add(Dense(32, activation = "relu"))
model.add(Dropout(0.2))
model.add(Dense(32, activation = "relu"))
model.add(Dropout(0.2))
model.add(Dense(32, activation = "relu"))
model.add(Dense(32, activation = "relu"))
model.add(Dense(32, activation="relu", kernel_regularizer=regularizers.l2(0.001)))
model.add(Dense(1))

model.compile(optimizer = "adam", loss = "mse")
```

```python
# Train model
history = model.fit(x_train, y_train, epochs = 500, batch_size = 32, validation_split = 0.2)
```

# Evaluation

```python
# Predictions
ypred_test = model.predict(x_test).flatten()
ypred_train = model.predict(x_train).flatten()
```

```python
# metrics
test_mse = mean_squared_error(y_test, ypred_test)
test_rmse = np.sqrt(test_mse)
test_r2 = r2_score(y_test, ypred_test)

print("Test_RMSE: ", test_rmse)
print("Test_R2 score: ",test_r2)

train_mse = mean_squared_error(y_train, ypred_train)
train_rmse = np.sqrt(train_mse)
train_r2 = r2_score(y_train, ypred_train)

print("Train_RMSE: ", train_rmse)
print("Train_R2 score: ",train_r2)
```

```python
# Training curve
pp.plot(history.history["loss"], label = "Training loss")
pp.plot(history.history["val_loss"], label = "Validation loss")
pp.legend()
pp.show()
```

```python
y_test_arr = np.array(y_test)
y_pred_arr = np.array(ypred_test)

# Sort by actual values for a smooth curve
sorted_index = np.argsort(y_test_arr)

x_actual = y_test_arr[sorted_index]
y_predicted = y_pred_arr[sorted_index]

pp.plot(x_actual, label = "Actual resale price", color = "green", marker = "*")
pp.plot(y_predicted, label = "Predicted resale price", color = "red", marker = ".",  alpha = 0.5)

pp.xlabel("Sorted Samples")
pp.ylabel("Predicted Resale Price")
pp.title("Actual vs Predicted Resale Price Curve")
pp.legend()
pp.show()
```

```python
pp.scatter(y_test, ypred_test, alpha = 0.5, edgecolors = "black", label = "Predictions")
min_val = min(y_test.min(), ypred_test.min())
max_val = max(y_test.max(), ypred_test.max())

pp.plot([min_val, max_val], [min_val, max_val], color = "red", label = "Min/Max Value Line")  # diagonal reference line

pp.xlabel("Actual Price")
pp.ylabel("Predicted Price")
pp.title("Actual vs Predicted Resale Price Scatter Plot")
pp.legend()
pp.show()
```

```python
lr=  LinearRegression()
lr.fit(x_train, y_train)
```

```python
ytrain_pred = lr.predict(x_train)
ytest_pred = lr.predict(x_test)
```

```python
lr_test_rmse = np.sqrt(mean_squared_error(y_test, ytest_pred))
lr_train_rmse = np.sqrt(mean_squared_error(y_train, ytrain_pred))

lr_test_r2 = r2_score(y_test, ytest_pred)
lr_train_r2 = r2_score(y_train, ytrain_pred)

print("Linear Regression Results")
print("Test RMSE:", lr_test_rmse)
print("Test R2:", lr_test_r2)
print("Train RMSE:", lr_train_rmse)
print("Train R2:", lr_train_r2)
```

```python
pp.scatter(y_test, ytest_pred, alpha=0.6, label="Linear Regression")
pp.scatter(y_test, ypred_test, alpha=0.6, label="ANN")

min_val = min(y_test.min(), ytest_pred.min(), ypred_test.min())
max_val = max(y_test.max(), ytest_pred.max(), ypred_test.max())

pp.plot([min_val, max_val], [min_val, max_val])

pp.xlabel("Actual Price")
pp.ylabel("Predicted Price")
pp.title("Linear & ANN Model Compare")
pp.legend()
pp.show()
```
